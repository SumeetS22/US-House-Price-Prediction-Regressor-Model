```python
replaced_data_values = dict({
    "Alley":"No_alley_acces",
    "BsmtQual":"No_Basement",
    "BsmtCond":"No_Basement",
    "BsmtExposure": "No_Basement",
   "BsmtFinType1": "No_Basement",
   "BsmtFinType2": "No_Basement",
   "FireplaceQu": "No_Fireplace",
   "GarageType": "No_Garage",
   "GarageFinish": "No_Garage",
   "GarageCond": "No_Garage",
   "GarageQual": "No_Garage",
   "PoolQC": "No_Pool",
   "MiscFeature": "Nothing",
   "Fence": "No_Fence",
   "MasVnrType": "Nothing"
})
```

```python
def replace_data(df):
    for idx in replaced_data_values:
        df[idx].fillna(replaced_data_values[idx], inplace=True)
    return df
```

```python
df_combined1 = replace_data(df_combined)
df_combined1.isnull().sum()
```

```python
categorical = df_combined.dtypes[df_combined.dtypes == 'object'].index
print(categorical)
```

```python
df_combined1[categorical].isna().sum()
```

```python
numerical = df_combined1.dtypes[df_combined.dtypes != 'object'].index
print(numerical)
```

```python
df_combined1[numerical].isna().sum()
```

```python
for i in categorical:
    df_combined1[i].fillna(df_combined1[i].mode(), inplace=True)
    
for i in numerical:
    df_combined1[i].fillna(df_combined1[i].median(), inplace=True)
```

```python
dictn = {}
for i in range(len(df_combined1.isnull().sum())):
    if (df_combined1.isnull().sum()[i])>0:
        dictn[f'{df_combined1.columns[i]}'] = df_combined1.isnull().sum()[i]
        
dictn
```

```python
df_combined2 = pd.get_dummies(df_combined1, drop_first=True)
df_combined2.head().style
```

```python
df_train.shape
```

```python
df_train_processed = df_combined2[:1460]
df_train_processed.shape
```

```python
# example of finding correlation
df_train_processed['MSSubClass'].corr(Sale_price)
```

```python
columns_to_maintain = []

for i in df_train_processed.columns:
    if abs(df_train_processed[i].corr(Sale_price))>0.40:
        columns_to_maintain.append(i)

columns_to_maintain
```

```python
df_train_processed[columns_to_maintain].corr()
```

```python
X = df_train_processed[columns_to_maintain]
y = Sale_price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train, y_train)
model1.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor()
model2.fit(X_train, y_train)
model2.score(X_test, y_test)
```

```python
test_data = df_combined2[1460:]
test_data.shape
```

```python
predictions = model1.predict(test_data[columns_to_maintain])
```

```python
pred = pd.DataFrame(predictions)
pred.rename(columns={0: 'SalePrice'}, inplace=True)
final = pd.concat([df_test.Id, pred], axis=1)
final.head()
```

```python
### Create sample submission file and submit

y_pred = pd.DataFrame(pred)
sub_df = pd.read_csv('/Users/animeshprasad/Desktop/Jupyter notebook/House Price regressor model/House Pricing by cmbning train and test/sample_submission.csv')
datasets = pd.concat([sub_df['Id'], y_pred], axis = 1)
datasets.columns = ['Id','SalePrice']
datasets.to_csv('sample_submission.csv', index = False)
```
